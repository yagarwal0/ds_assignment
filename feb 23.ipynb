# Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

> read_csv() - function is used to read from the csv file

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

> head() - function returns the first 5 rows

```python
df.head()
```

> tail() - function returns the last 5 rows 

```python
df.tail()
```

> sort_values() - function is used to sort the values along any axis

```python
df.sort_values('Fare')
```

> iloc() - function gives the subset of data from the complete data

```python
df.iloc[8:98, 2:8]
```

# Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
name = pd.Series(['yash','priya','monica'])
age = pd.Series([23,98,99])
sex = pd.Series(['m','f','m'])
```

```python
df = pd.DataFrame({
    'name' : name,
    'age' : age,
    'sex' : sex})
```
