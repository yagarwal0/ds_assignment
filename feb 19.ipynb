# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function. 

In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint based on the endpoint name and any additional arguments or parameters required by the endpoint.

```python
@app.route('/sport/<name>/<team>')
def game(name, team):

    if name == 'cricket':
        return redirect(url_for('cricket', favteam=team))
    elif name == 'baseball':
        return redirect(url_for('baseball', favteam=team))
    elif name == 'carrom':
        return redirect(url_for('carrom', favteam=team))
    else:
        return render_template('workinprogress.html',
                               sportname=name)
```
