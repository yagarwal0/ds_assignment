# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter plot is a graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
plt.scatter(x,y)
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

subplot() function is the part of matplotlib library and it is used to create several plots within a figure

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig = plt.figure()
f1 = fig.add_subplot(2,2,1)
f1.plot(x1, y1)
f2 = fig.add_subplot(2,2,2)
f2.plot(x2, y2)
f3 = fig.add_subplot(2,2,3)
f3.plot(x3, y3)
f4 = fig.add_subplot(2,2,4)
f4.plot(x4, y4)
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot, also known as a bar chart or bar graph, is a common type of data visualization used to display categorical data. It represents data using rectangular bars, where the length or height of each bar corresponds to the magnitude or frequency of the data being represented.

It is mainly used to 
* make comparision of of categories
* visualizing the frequencies

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.barh(company, profit)
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Box plot is the plot which displays the median, quartiles, and potential outliers in a concise manner.

It is used for:-
* Detection of outliers
* Summary of distribution

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.show()
```

```python
plt.boxplot(box2)
plt.show()
```
