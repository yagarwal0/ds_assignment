# Q1) Explain why we have to use the Exception class while creating a custom exception

Using the Exception class as the base class for custom exceptions ensures that your custom exception inherits important properties and behaviors from the Exception class, making it compatible with the existing exception handling mechanisms and conventions of the programming language.

# Q2. Write a python program to print Python Exception Hierarchy.

```python
import inspect
```

```python
def hierarchy(cls, ind = 1):
    print('-'*ind,'>', cls.__name__)
    for i in cls.__subclasses__():
        hierarchy(i, ind+3)
```

```python
hierarchy(Exception)
```
